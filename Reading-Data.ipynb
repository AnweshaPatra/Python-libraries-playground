Concatenate all the files in a single csv file

```python
data_frame_list = []

for directory in glob.glob("/Users/annie/Documents/GitHub/Python-libraries-playground/datasets/multi-directory/*"):
    for files in glob.glob(directory + "/*"):
        data_frame_list.append(pd.read_csv(files))

data_frame_list
```

```python
final_data = pd.concat(data_frame_list)
```

```python
final_data.head()
```

```python
final_data.shape
```

```python
final_data.to_csv("/Users/annie/Documents/GitHub/Python-libraries-playground/datasets/merged_big_mart_sales.csv", index = False)
```
